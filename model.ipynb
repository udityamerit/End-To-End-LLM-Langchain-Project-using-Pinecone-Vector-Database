```python
import os

def read_doc(folder_path):
    docs = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt") or filename.endswith(".pdf") or filename.endswith(".docx"):
            file_path = os.path.join(folder_path, filename)
            docs.append(file_path)
    return docs

user_input = input("Enter the path to your document folder (e.g. documents/): ")

doc = read_doc(user_input)
print("Documents Loaded:", doc)
```

```python
# Divide the docs into chunks

def chunk_data(docs, chunk_size =800, chunk_overlap=50):
        text_splitter= RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
        doc= text_splitter.split_documents(docs)
        return docs
```

```python
documents = chunk_data(docs=doc)
documents 
```
